# Intro to Data Science

Data Science is about drawing useful insights, or conclusions from the data through **exploration**, **prediction**, and **inference**.

* Exploration involves identifying patterns in information. Tools: **descriptive statistic**, **visualization**.
* Prediction involves using the data we know to make informed guesses about unseen data. Tools: **machine learning**.
* Inference involves quantifying the degree of certainty: will the patterns that we found in the data also appear in new observations?, how accurate are our predictions?. Tools: **statistical tests** and **models**.

## Descriptive statistic & visualization

In this example, we will explore statistics for two classic novels: *The Adventures of Huckleberry Finn* by Mark Twain, and *Little Women* by Louisa May Alcott.

```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("whitegrid")
```

We read the text of both books into list of chapters

```
from urllib.request import urlopen 
import re


def read_url(url): 
    return re.sub('\\s+', ' ', urlopen(url).read().decode())

huck_finn_text = read_url('https://raw.githubusercontent.com/dhminh1024/practice_datasets/master/huck_finn.txt')
huck_finn_chapters = pd.DataFrame({'Chapters':huck_finn_text.split('CHAPTER ')[44:]})

little_women_text = read_url('https://raw.githubusercontent.com/dhminh1024/practice_datasets/master/little_women.txt')
little_women_chapters = pd.DataFrame({'Chapters':little_women_text.split('CHAPTER ')[1:]})


```

```
little_women_chapters.head(10)
```

***The Adventures of Huckleberry Finn*** describes a journey of Huck, Jim and Tom Sawyer. We can visualize how many times these characters have been mentioned at any point of the book:

```
counts = pd.DataFrame({
    'Jim': huck_finn_chapters['Chapters'].str.count('Jim'),
    'Tom': huck_finn_chapters['Chapters'].str.count('Tom'),
    'Huck': huck_finn_chapters['Chapters'].str.count('Huck'),
})
```

```
counts = counts.cumsum()

fig, ax = plt.subplots(figsize=(10, 6))
counts.plot(ax=ax)
ax.set(title="Cumulative Number of Times Each Name Appears", xlabel='Chapter')
plt.show()
```

```
#@title What can you tell? {display-mode: "form"}


# We can see that:
# * Jim is a central character by the large number of times his name appears
# * After Chapter 30, Tom joins the adventures. His curve and Jim's rise sharply at that point.
# * As for Huck, his name hardly appears at all, because he is the narrator
```

***Little Women*** is a story of four sisters growing up together during the civil war.

```
counts = pd.DataFrame({
    'Amy': little_women_chapters['Chapters'].str.count('Amy'),
    'Beth': little_women_chapters['Chapters'].str.count('Beth'),
    'Jo': little_women_chapters['Chapters'].str.count('Jo'),
    'Meg': little_women_chapters['Chapters'].str.count('Meg'),
    'Laurie': little_women_chapters['Chapters'].str.count('Laurie'),
})

counts = counts.cumsum()

fig, ax = plt.subplots(figsize=(10, 6))
counts.plot(ax=ax)
ax.set(title="Cumulative Number of Times Each Name Appears", xlabel='Chapter')
plt.show()
```
